```python
import pandas as pd
import matplotlib.pyplot as plt

# Read file
data = pd.read_csv('v1_herz_1.csv')

# Select required columns
heard_data = data[['X', 'Y', 'Z']]

# Assign variables
time = heard_data.index # Timeline
X = heard_data['X']     # X-Values
Y = heard_data['Y']     # Y-Values
Z = heard_data['Z']     # Z-Values

# Create a figure and set its size to 15x5 in.
fig, ax = plt.subplots(figsize=(15,5))

# Plot a line chart
ax.plot(time, X)
ax.plot(time, Y)
ax.plot(time, Z)

# Decrease the density of ticks on x-axis
#ax.set_xticks(date[::12])

# Add title and axes labels
ax.set_title('Results heard beat')
ax.set_xlabel('Time (x10 ms)')
ax.set_ylabel('Values')

# Show plot
plt.show()
```
